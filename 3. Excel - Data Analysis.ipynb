### II. ANSWERS

##### 1. OVERVIEW TABLE

Once the entire table of data has been presented, we point out those that are useful for answering the three questions presented.

In this case, we will mainly focus on the "Claim ID", "Location" and "Time to close".

#### 2. ANALYSIS PLAN

*2.1 Tool used*

In order to answer the questions we used the spreadsheet tool from Microsoft Office - Excel, through which pivot tables and graphs were created to provide the desired information.

*2.2 Data processing*
  
First of all, we processed the data that could be misleading for the subsequent data analysis. In this case, it has only been necessary to focus on the "Time to Close", where Claim number 58 has a negative value, which is inconsistent and has been eliminated. Otherwise, all data are ready to be analysed.

#### 3. VISUALIZATIONS

The first question refers to the number of complaints filed by location. In this case we need the "Location" data and the "Claim ID".

For this, we have made a dynamic table in excel (TABLE 1), where the different locations create different rows and the "Claim ID" values are added. In this way, the total count of Claims per location is automatically generated.


For the second question, a table has been created with the values of "Time to close" and "Claim ID", forming 7 ranges of time values, taking as reference the smallest and the largest value referring to the time to close claims. 

With this data, excel creates a mathematical formula that distributes the time ranges equally according to the number of complaints. It can be seen that the majority of complaints are closed during the time range between 0 to 500 days (Figure 2) having as a result a right skewed distribnution.

During the data processing of this exercise we found, in the values (in days) of "Time to close", a negative value, which does not make sense because no complaint can be resolved in a negative number of days. Therefore, it has been decided to ignore the negative sign and to treat the data in absolute values.

For the third and last question, we need the results of the first question (number of claims per location) and the "Time to close" for each complaint. With this, we generate another pivot table by adding the time to close values to the existing one.

As a result, the total times to close of the complaints by location are added up (Figure 3). Finally, we average each result obtained by location (average between the sum of the time to close by location and the sum of complaints by location). The command used is: "=AVERAGE(Count_of_ClaimID;Count_of_Time_to_Close)".

#### 4. CONCLUSSIONS

With the data obtained, it can be concluded that the majority of complaints processes are concluded in a time range of between 29 to 1069 days and that, considering the 4 locations studied, the area of Sao Luis has the most complaints and takes the longest time to close complaints.

Finally, the correlation between the time to close complaints and the number of complaints should be studied in future lines of research. In order to find out whether Sao Luis is the area that takes the longest time to resolve its complaints for reasons of quantity or for reasons of efficiency.
